```python
datos = pd.read_csv('alquiler.csv', sep = ';')
datos.head(10)
```

```python
list(datos['Tipo'].drop_duplicates())
```

```python
residencial = ['Habitación',
 'Casa',
 'Local comercial',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa']
residencial
```

```python
datos.head(10)
```

```python
datos_residencial = datos[seleccion]
datos_residencial
```
